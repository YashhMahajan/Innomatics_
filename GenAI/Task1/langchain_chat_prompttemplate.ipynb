### ChatPromptTemplate System

```python
def get_chat_prompt(role, topic):
    
    if role == "teacher":
        system_template = "You are a helpful teacher."
        human_template = "Teach me about {topic}"
        
    elif role == "interviewer":
        system_template = "You are an interviewer."
        human_template = "Ask me questions about {topic}"
        
    elif role == "motivator":
        system_template = "You are a motivational speaker."
        human_template = "Motivate me to learn {topic}"
        
    else:
        raise ValueError("Invalid role")

    chat_prompt = ChatPromptTemplate.from_messages([
        SystemMessagePromptTemplate.from_template(system_template),
        HumanMessagePromptTemplate.from_template(human_template)
    ])
    
    return chat_prompt.format_messages(topic=topic)

# Example
messages = get_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

### Input Validation Layer

```python
def validate_inputs(audience, tone):
    
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]
    
    if audience not in valid_audience:
        print("Invalid audience. Defaulting to 'beginner'")
        audience = "beginner"
        
    if tone not in valid_tone:
        print("Invalid tone. Defaulting to 'formal'")
        tone = "formal"
        
    return audience, tone

# Test
print(validate_inputs("kids", "fun"))
print(validate_inputs("expert", "serious"))
```

### Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    
    # Validate inputs
    audience, tone = validate_inputs(audience, tone)
    
    # Style templates
    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }
    
    if style not in templates:
        raise ValueError("Invalid style")
    
    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )
    
    return prompt_template.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# ONE reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Different inputs
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("Data Science", "expert", "casual"),
    ("Machine Learning", "beginner", "formal"),
    ("Deep Learning", "expert", "fun")
]

for topic, audience, tone in inputs:
    print(reusable_template.format(topic=topic, audience=audience, tone=tone))
```
