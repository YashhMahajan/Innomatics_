#### Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    return False

while True:
    amount = int(input("Enter recharge amount: "))
    if validate_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Invalid recharge. Try again.")
```

#### Problem 2: Inventory Reorder Alert System


```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, "- Reorder Alert")
        else:
            print(item, "- Stock OK")

products = {
    "Pen": 10,
    "Notebook": 25,
    "Pencil": 12,
    "Eraser": 30
}

check_inventory(products)
```

#### problem 3: Student Result Processing System


```python
def process_result(marks):
    total = 0
    for m in marks:
        total += m
    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 70, 45, 55]
result = process_result(marks)
print(result)
```

#### Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    if peak.lower() == "yes":
        fare += fare * 0.25
    return fare

while True:
    try:
        distance = float(input("Enter distance in km: "))
        if distance < 0:
            print("Distance cannot be negative. Try again.")
            continue

        peak = input("Is it peak hour? (yes/no): ").lower()
        if peak not in ["yes", "no"]:
            print("Invalid input for peak hour. Try again.")
            continue

        fare = calculate_fare(distance, peak)
        print("Total Fare:", fare)
        break

    except:
        print("Invalid input. Try again.")
```

#### Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
result = check_eligibility(attendance)
print(result)
```

#### Problem 6: Password Strength Checker

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter password: ")
result = check_password(password)
print(result)
```
