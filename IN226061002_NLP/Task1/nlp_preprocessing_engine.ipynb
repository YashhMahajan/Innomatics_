### Download Resources

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
# Download required datasets
nltk.download('stopwords')
nltk.download('wordnet')
```

### Initialize Tools

```python
# Initialize stopwords and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Sample text
text = "I Love this Movie!!"

print(f"Original text: {text}")
```

### Preprocessing Function (Advanced + Corrected)

```python
def preprocess_text(text):
    """
    Advanced NLP preprocessing function
    """
    
    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return []
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove special characters / emojis
    text = re.sub(r'[^\w\s]', ' ', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove short words except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    # Remove stopwords (but keep 'no', 'not')
    tokens = [word for word in tokens if word not in stop_words or word in ['no', 'not']]
    
    # Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return tokens


# Run function
processed = preprocess_text(text)

print(f"Processed text: {processed}")
```

### Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)
    
    results.append((sentence, tokens, clean_sentence))
    
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sentence)
    print("-" * 50)
```

### Token Analytics

```python
def analyze_tokens(tokens):
    if not tokens:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total
    
    return total, unique, avg_len


for sentence, tokens, clean in results:
    total, unique, avg = analyze_tokens(tokens)
    
    print("Sentence:", clean)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
    print("-" * 50)
```

### Frequency Analysis

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
#full pipeline 
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens = preprocess_text(text)
        clean = " ".join(tokens)
        
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


pipeline_output = full_pipeline(test_sentences)

print(pipeline_output)
```

### Error Handling
